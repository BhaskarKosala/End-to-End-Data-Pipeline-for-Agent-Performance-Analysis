Task 1 - Data Ingestion and Validation

```python
#Reading input files
call_logs_df = pd.read_csv(r'G:\Bhaskar\DPDzero Data Ops Assignment\call_logs.csv')

agent_roster_df = pd.read_csv(r'G:\Bhaskar\DPDzero Data Ops Assignment\agent_roster.csv')

disposition_summary_df = pd.read_csv(r'G:\Bhaskar\DPDzero Data Ops Assignment\disposition_summary.csv')
```

```python
#Ensuring call_date, agent_id, and org_id are present and correctly formatted
call_logs_df['agent_id'],call_logs_df['org_id'],agent_roster_df['agent_id'],agent_roster_df['org_id'],disposition_summary_df['agent_id'],disposition_summary_df['org_id'],disposition_summary_df['call_date']
```

```python
#Flag missing or duplicate entries
#Creating duplicate_flag, missing_flag fields into each dataframe
#duplicate_flag will be set to True if that record is a duplicated record
#missing_flag will be set to True if any of the values in that record are missing
#call_logs_df
call_logs_df['duplicate_flag']=call_logs_df.duplicated()
call_logs_df['missing_flag']=call_logs_df.isna().any(axis=1)

#agent_roster_df
agent_roster_df['duplicate_flag']=agent_roster_df.duplicated()
agent_roster_df['missing_flag']=agent_roster_df.isna().any(axis=1)

#disposition_summary_df
disposition_summary_df['duplicate_flag']=disposition_summary_df.duplicated()
disposition_summary_df['missing_flag']=disposition_summary_df.isna().any(axis=1)
```

Task 2 - Join Logic

```python
#Merge the datasets using agent_id, org_id, and call_date.

#Merging the call_logs_df & agent_roster_df daatasets using 'agent_id','org_id' fields where call_logs_df being the driver
merged_df=pd.merge(call_logs_df,agent_roster_df, on = ('agent_id','org_id') , how = 'left')

#Merging the above merged dataframe with disposition_summary_df using 'agent_id','org_id' and 'call_date' fields
merged_df=pd.merge(merged_df,disposition_summary_df, on = ('agent_id','org_id','call_date'), how = 'left')
```

```python
print('Primary dataset count:',len(call_logs_df))
print('Merged dataset count:',len(merged_df))
```

Task 3 - Feature Engineering

```python
#For each agent on each date, compute Total Calls Made
call_counts = merged_df.groupby(['agent_id','call_date']).size().reset_index(name='total_calls_made')
call_counts
```

```python
#For each agent on each date, compute Unique Loans Contacted
unique_loans = merged_df.groupby(['agent_id','call_date'])['installment_id'].nunique().reset_index(name = 'unique_loans_contacted')
unique_loans
```

```python
#For each agent on each date, compute Connect Rate = Completed Calls / Total Calls
completed_calls_df = merged_df[merged_df['status'] == 'completed'].groupby(['agent_id','call_date'])['status'].count().reset_index(name='completed_calls')
completed_calls_df
```

```python
connect_rate_df = pd.merge(call_counts,completed_calls_df, on = ('agent_id','call_date'), how = 'left')
connect_rate_df['connect_rate'] = connect_rate_df['completed_calls'] / connect_rate_df['total_calls_made']
connect_rate_df
```

```python
#Avg Call Duration (in minutes)
avg_call_duration_df = merged_df.groupby(['agent_id','call_date'])['duration'].mean().reset_index(name='avg_call_duration')
avg_call_duration_df['avg_call_duration_minutes'] = avg_call_duration_df['avg_call_duration'].apply(lambda x: f"{int(x)}:{int((x - int(x)) * 60) :02d}")
avg_call_duration_df
```

```python
#Presence (1 if login_time exists, else 0)
presence_grouping = merged_df.groupby(['agent_id','call_date']).agg(is_present = ('login_time','max'))
presence_grouping['presence'] = 0
for i in range(len(presence_grouping)):
    curr_value = (presence_grouping.iloc[i]['is_present'])
    if str(curr_value) == 'nan':
        presence_grouping['presence'][i] = 0
    else:
        #presence_grouping.loc[i,'presence'] = 1
        presence_grouping['presence'][i] = 1
```

```python
top_performer = connect_rate_df[connect_rate_df['connect_rate'] == max(connect_rate_df['connect_rate'])]['agent_id']
max_connect_rate = round(max(connect_rate_df['connect_rate']) * 100,2)                    
```

```python
active_agents = len(presence_grouping[presence_grouping['presence'] == 1])
```

```python
avg_duration_top_performer = avg_call_duration_df[avg_call_duration_df['agent_id'] == top_performer.iloc[0]]['avg_call_duration_minutes']
```
